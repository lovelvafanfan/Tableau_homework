```python
import pandas as pd
import numpy as ny
```

```python
# read all 12 months' city bike trip data
df1 =  pd.read_csv("2016rawdata/201601.csv")
df2 =  pd.read_csv("2016rawdata/201602.csv")
df3 =  pd.read_csv("2016rawdata/201603.csv")
df4 =  pd.read_csv("2016rawdata/201604.csv")
df5 =  pd.read_csv("2016rawdata/201605.csv")
df6 =  pd.read_csv("2016rawdata/201606.csv")
df7 =  pd.read_csv("2016rawdata/201607.csv")
df8 =  pd.read_csv("2016rawdata/201608.csv")
df9 =  pd.read_csv("2016rawdata/201609.csv")
df10 =  pd.read_csv("2016rawdata/201610.csv")
df11 =  pd.read_csv("2016rawdata/201611.csv")
df12 =  pd.read_csv("2016rawdata/201612.csv")
```

```python
# Concatenate all 12 months data
data2016 = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12])
data2016
```

```python
# export new csv file
data2016.to_csv('citibike_2016.csv') 
```
